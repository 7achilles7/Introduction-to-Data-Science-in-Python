```python
import matplotlib.pyplot as plt
```

```python
x1 = [1,5,10]
y1 = [6,8,3]
x2 = [2,7,11]
y2 = [2,8,4]
plt.plot(x1,y1,label = 'first one')
plt.plot(x2,y2, label = 'second one')
plt.plot([],[],color = 'm',label = 'fake')
plt.xlabel('x axis')
plt.ylabel('y label')
plt.title('First Graph\n')
plt.legend()
plt.show()
```

```python
slices = [10,4,2,7,3]

plt.pie(slices,
        labels = ['car','bike','bus','train','aeroplane'],
        colors = ['r','c','m','b','k'],
        startangle = 90,
        explode = (0,0.1,0,0,0),
        shadow = True,
        autopct = '%1.1f%%')
plt.show()
```

```python
days = [1,2,3,4,5]

eating = [2,2,2,2,2]
sleeping = [8,9,9,7,8]
office = [7,7,7,6,6]

plt.stackplot(days,eating,sleeping,office)

plt.legend()
plt.show()
```
